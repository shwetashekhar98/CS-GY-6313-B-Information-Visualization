```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import wbdata
import datetime

# Fetch data from World Bank API
def fetch_wbdata():
    try:
        indicators = {"SI.POV.GINI": "Gini Index"}
        start_date = datetime.datetime(1990, 1, 1)
        end_date = datetime.datetime(2020, 12, 31)

        data = wbdata.get_dataframe(indicators, date=(start_date, end_date))
        data.reset_index(inplace=True)
        data.rename(columns={"country": "Country", "date": "Year"}, inplace=True)

        if data.empty:
            raise ValueError("No data fetched from World Bank API.")
        return data
    except Exception as e:
        print(f"Error fetching data: {e}")
        raise

# Process the data
def process_data(data, start_year=2000, end_year=2020):
    try:
        available_years = data["Year"].unique()
        if start_year not in available_years or end_year not in available_years:
            start_year, end_year = available_years.min(), available_years.max()

        gini_data = data[data["Year"].isin([start_year, end_year])].pivot(
            index="Country", columns="Year", values="Gini Index"
        ).reset_index()
        gini_data.columns.name = None
        gini_data = gini_data.rename(columns={start_year: "start", end_year: "end"})
        gini_data = gini_data.dropna(subset=["start", "end"])
        gini_data = gini_data.reset_index(drop=True)
        gini_data["change"] = gini_data["end"] - gini_data["start"]
        gini_data["color"] = np.where(gini_data["change"] > 0, "#d62728", "#2ca02c")  # Red for positive, green for negative
        return gini_data, start_year, end_year
    except Exception as e:
        print(f"Error processing data: {e}")
        raise

# Fetch and process data
try:
    raw_data = fetch_wbdata()
    gini_data, start_year, end_year = process_data(raw_data)
except Exception as e:
    print(f"Critical Error: {e}")
    exit()

# Pre-filter data for positive and negative changes
positive_data = gini_data[gini_data["change"] > 0]
negative_data = gini_data[gini_data["change"] <= 0]

# Initialize Dash app
app = dash.Dash(__name__)

# Predefined Figure
def create_initial_figure():
    fig = go.Figure()

    # Add all data
    angles_all = np.linspace(0, 360, len(gini_data), endpoint=False)
    for i, row in gini_data.iterrows():
        fig.add_trace(go.Scatterpolar(
            r=[row["start"], row["end"]],
            theta=[angles_all[i], angles_all[i]],
            mode="lines+markers+text",
            name=row["Country"],
            marker=dict(size=12, symbol="circle", color=row["color"]),
            line=dict(color=row["color"], width=3),
            text=[f"{row['start']:.1f}", f"{row['end']:.1f}"],
            textposition=["bottom center", "top center"],
            visible=True  # Initially visible
        ))

    # Add positive data
    angles_positive = np.linspace(0, 360, len(positive_data), endpoint=False)
    for idx, (i, row) in enumerate(positive_data.iterrows()):
        fig.add_trace(go.Scatterpolar(
            r=[row["start"], row["end"]],
            theta=[angles_positive[idx], angles_positive[idx]],
            mode="lines+markers+text",
            name=row["Country"],
            marker=dict(size=12, symbol="circle", color="#d62728"),
            line=dict(color="#d62728", width=3),
            text=[f"{row['start']:.1f}", f"{row['end']:.1f}"],
            textposition=["bottom center", "top center"],
            visible=False  # Initially hidden
        ))

    # Add negative data
    angles_negative = np.linspace(0, 360, len(negative_data), endpoint=False)
    for idx, (i, row) in enumerate(negative_data.iterrows()):
        fig.add_trace(go.Scatterpolar(
            r=[row["start"], row["end"]],
            theta=[angles_negative[idx], angles_negative[idx]],
            mode="lines+markers+text",
            name=row["Country"],
            marker=dict(size=12, symbol="circle", color="#2ca02c"),
            line=dict(color="#2ca02c", width=3),
            text=[f"{row['start']:.1f}", f"{row['end']:.1f}"],
            textposition=["bottom center", "top center"],
            visible=False  # Initially hidden
        ))

    # Layout
    fig.update_layout(
        polar=dict(
            bgcolor="#f7f7f7",
            angularaxis=dict(
                tickmode="array",
                tickvals=angles_all,
                ticktext=gini_data["Country"],
                rotation=90,
                direction="clockwise",
                tickfont=dict(size=12, color="black")
            ),
            radialaxis=dict(
                title=dict(text="Gini Index", font=dict(size=14, color="black")),
                range=[0, max(gini_data[["start", "end"]].values.max(), 80)],
                gridcolor="lightgray",
                gridwidth=0.5
            ),
        ),
        title=dict(
            text=f"<b>Radial Slope Chart:</b> Change in Gini Index ({start_year} to {end_year})",
            font=dict(size=18),
            x=0.5, xanchor="center", y=0.95
        ),
        showlegend=True,
        margin=dict(l=50, r=50, t=120, b=50)
    )

    return fig

# App layout
app.layout = html.Div([
    html.H1("Interactive Radial Slope Chart: Gini Index", style={'textAlign': 'center'}),
    html.Div([
        html.Label("Select Filter:"),
        dcc.Dropdown(
            id="filter-dropdown",
            options=[
                {"label": "All Countries", "value": "all"},
                {"label": "Positive Change", "value": "positive"},
                {"label": "Negative Change", "value": "negative"},
            ],
            value="all",
            style={"width": "50%"}
        )
    ], style={'textAlign': 'center', 'margin-bottom': '20px'}),
    dcc.Graph(id="radial-chart", figure=create_initial_figure(), style={"height": "800px"})
])

# Callback to toggle visibility
@app.callback(
    Output("radial-chart", "figure"),
    [Input("filter-dropdown", "value")]
)
def update_visibility(selected_filter):
    fig = create_initial_figure()
    # Set visibility based on the selected filter
    if selected_filter == "positive":
        # Hide all traces, then show only positive traces
        for trace in fig.data:
            trace.visible = False
        for i in range(len(gini_data), len(gini_data) + len(positive_data)):
            fig.data[i].visible = True
    elif selected_filter == "negative":
        # Hide all traces, then show only negative traces
        for trace in fig.data:
            trace.visible = False
        for i in range(len(gini_data) + len(positive_data), len(fig.data)):
            fig.data[i].visible = True
    else:  # Show all countries
        for trace in fig.data:
            trace.visible = True
    return fig

# Run the app
if __name__ == "__main__":
    app.run_server(debug=True)
```
